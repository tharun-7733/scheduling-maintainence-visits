Importing Dataset from Kaggle

```python
import kagglehub

# Download latest version
path = kagglehub.dataset_download("stratographics/predictive-maintenance-equipment-dataset-for-ml")

print("Path to dataset files:", path)
```

```python
import os

# List all files in the dataset directory
print(os.listdir(path))
```

Exploring the loaded data and handlening targeted data

```python
import pandas as pd
import os
train_path = os.path.join(path,"PM DATA Training.xlsx")
test_path = os.path.join(path,"PM Data Testing.xlsx")

train_df = pd.read_excel(train_path)
test_df = pd.read_excel(test_path)

print("Traing data loaded:",train_df.shape)
print("Testing data loaded:",test_df.shape)
```
